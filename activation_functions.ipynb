```python
import torch
import matplotlib.pyplot as plt
```

# ReLU

# torch.maximum(0, x)

```python
#relu
#very straight forward, just clip when a value is less than 0
x = torch.range(-10, 10)
```

```python
x
```

```python
plt.plot(x, torch.maximum(torch.tensor(0, dtype=torch.float64), x))
```

#sigmoid [0, 1]
# $\frac{1}{1+e^{-x}}$

```python
#lets break this down
x = torch.arange(-10, 10)
plt.plot(x, x)
```

```python
#e^x
plt.plot(x, torch.exp(x))
```

```python
#e^-x
plt.plot(x, torch.exp(-x))
```

```python
#plot 1/e^x
plt.plot(x, 1/(torch.exp(x)))
```

```python
#plot 1/e^x
plt.plot(x, 1/(torch.exp(-x)))
```

```python
#plot 1/(1 + e^-x)
plt.plot(x, 1/(1+ torch.exp(-x)))
```

# Tanh [-1, 1]
# $ \frac{e^x-e^{-x}}{e^x + e^{-x}} $

```python
plt.plot(x, (torch.exp(x) - torch.exp(-x))/(torch.exp(x) + torch.exp(-x)))
```
